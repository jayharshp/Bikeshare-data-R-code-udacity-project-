### Question 1


**What is the top 50 end stations used most frequently in different cities? Which one is the most common end station per city?**

```python



es = ny[6]

#we will make a data frame that has a table with frequency of each end station

a = as.data.frame(table(es))
#a

# now we will extract top 50 end stations by rearranging data in descending order

desc_freq = a[with(a, order(-Freq)), ] 

#desc_freq

b = head(desc_freq, 50)

#b

summary(b)

library(ggplot2)

g = ggplot(b, aes(x = es, y = Freq))

g + geom_bar(stat="identity", fill = "steelblue") + coord_flip () + labs (title = "Top 50 End Stations in NY", x = "Station Name", y = "Count") + scale_y_continuous(limit = c(0, 600)) + geom_text(aes(label=Freq), hjust = 1.5, color = "white", size = 3) + theme_minimal() 








esw = wash[6]

#esw

#we will make a data frame that has a table with frequency of each end station

aw = as.data.frame(table(esw))


names(aw)[2] = "Freq_w"

#aw

# now we will extract top 50 end stations

desc_freq_w = aw[with(aw, order(-Freq_w)), ] 

#desc_freq_w

bw = head(desc_freq_w, 50)

#bw
summary(bw)

gw = ggplot(bw, aes(x = esw, y = Freq_w))

gw + geom_bar(stat="identity", fill = "steelblue") + coord_flip () + labs (title = "Top 50 End Stations in Washington", x = "Station Name", y = "Count") + scale_y_continuous(limit = c(0, 2000)) + geom_text(aes(label=Freq_w), hjust = 1.5, color = "white", size = 3) + theme_minimal() 




esc = chi[6]

#esc

#we will make a data frame that has a table with frequency of each end station

ac = as.data.frame(table(esc))


names(ac)[2] = "Freq_c"

#ac

# now we will extract top 50 end stations

desc_freq_c = ac[with(ac, order(-Freq_c)), ] 

#desc_freq_c

bc = head(desc_freq_c, 50)

#bc

summary(bc)

gc = ggplot(bc, aes(x = esc, y = Freq_c))

gc + geom_bar(stat="identity", fill = "steelblue") + coord_flip () + labs (title = "Top 50 End Stations in Chicago", x = "Station Name", y = "Count") + scale_y_continuous(limit = c(0, 250)) + geom_text(aes(label=Freq_c), hjust = 1.5, color = "white", size = 3) + theme_minimal() 



```

### Question 3

**What is the relation between birth-year range of users and the count for bike rentals in Ney York?**

```python
# Your solution code goes here

# we will extract the column of interest here firstly.

year = ny$Birth.Year

# as the data has NA (missing values) we will omit them

year_biju = na.omit(year)


# run summary to roughly estimate the min and max of birth year

summary(year_biju)

#head(year_biju, 100)


# we will construct for loop to count users in every 50 years age bracket. 

count = 0
count1 = 0
count2 = 0
count3 = 0



for (val in year_biju) {
    if (val >= 1850 & val <= 1899)
        {
        count = count + 1;
        }
    else if (val >= 1900 & val <= 1949)
        {
        count1 = count1 + 1;
    }
    else if (val >= 1950 & val <= 1999)
        {
        count2 = count2 + 1;
    }
    else 
        {
        count3 = count3 + 1;
    }
        
}

#count
#count1
#count2
#count3

age_df = data.frame(Year_range = c('1850-1899', '1900-1949', '1950-1999', '2000-Present'), Count = c(count, count1, count2, count3))

# plot the range of birth year against count

library (ggplot2)
gc = ggplot(age_df, aes(x = Year_range, y = Count))

gc + geom_bar(stat="identity", fill = "black") +
labs (title = "Age range for bike rental user in NY", x = "Age Range", y = "Count")



```

**From the summary data, we can say that the birth-year of users range from 1885-2001. 48919 users are from the birth-year between 1950 and 1999. This is the highest user count birth-year range.
